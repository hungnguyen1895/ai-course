```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

**Minimum Point**

```
def f(x, y):
    return (x**2 + y**2);

x = np.linspace(-6, 6, 20)
y = np.linspace(-6, 6, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(5,30)

ax = plt.gca()

# plot point
point  = np.array([0, 0, 0])
ax.scatter(point[0], point[1], point[2], color='red')

# draw the gradients
fig, ax = plt.subplots()
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
q = ax.quiver(X, Y, U, V, 5.5)
ax.quiverkey(q, X=0, Y=0, U=0,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```

**Maximum Point**

```
def f(x, y):
    return (-x**2 + -y**2);

x = np.linspace(-6, 6, 20)
y = np.linspace(-6, 6, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(5,30)

ax = plt.gca()

# plot point
point  = np.array([0, 0, 0])
ax.scatter(point[0], point[1], point[2], color='red')

# draw the gradients
fig, ax = plt.subplots()
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
q = ax.quiver(X, Y, U, V, 5.5)
ax.quiverkey(q, X=0, Y=0, U=0,
             label='Quiver key, length = 10', labelpos='E')


plt.show()
```
